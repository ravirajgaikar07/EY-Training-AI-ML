```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
total_rev_per_product=df.groupby("Product")["Revenue"].sum()
```

```python
total_rev_per_product
```

```python
average_units_per_product=df.groupby("Product")["Units_Sold"].mean()
average_units_per_product
```

Visualization 1 – Line Chart
Plot the Monthly Revenue Trend for each product (Jan–Mar).

```python
for product in df["Product"].unique():
    product_data=df[df["Product"]==product]
    plt.plot(product_data["Month"],product_data["Revenue"],marker="o",label=product)
plt.title("Monthly Revenue Trend for each product")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

Visualization 2 – Bar Chart
Plot a bar chart of total revenue per product.

```python
plt.bar(df["Product"],df["Revenue"])
plt.title("Total revenue per product")
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.show()
```

Visualization 3 – Pie Chart
Plot a pie chart showing the contribution of each product to total
revenue.

```python
total_rev_per_product.plot(kind='pie',autopct="%1.1f%%",startangle=50)
plt.title("Contribution of each product to total revenue")
plt.show()
```
